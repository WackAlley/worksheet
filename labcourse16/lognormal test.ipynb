```python
# testing some values to make a good estimation for the initial mcmc prior

def lognormal(ch,mu1,sig1): 
    return np.divide(1,ch*sig1*np.sqrt(2*np.pi)) * np.exp(-np.divide(np.power(np.log(ch)-mu1,2),2*sig1**2))

x = np.linspace(-1000, 2400, 1000)
mu = 8
mu2 = 7
sig = 2.88
sig2= 2.3
amp = 5.54e4
amp2 = 5e4
plt.plot(x,amp*lognormal(x-30,mu,sig), color='k')
plt.plot(x,amp2*lognormal(x,mu2,sig2))
# plt.plot(x,120*np.exp(-x**2/300**2), color='g')
# plt.plot(x,120*x*np.exp(-x/450), color='y')
# plt.plot(x,amp2*np.divide(1,sig2*np.sqrt(2*np.pi)) * np.exp(-(x-mu2)**2 / (2*sig2**2)), color='k' )
# plt.axvline(np.exp(mu2-(sig2**2)), color='r')
# plt.plot(x, 50/x**1.3)
plt.axvline(110, color='r')
plt.show()
```
