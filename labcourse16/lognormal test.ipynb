```python
# testing some values to make a good estimation for the initial mcmc prio

def lognormal(ch,mu1,sig1): 
    return np.divide(1,ch*sig1*np.sqrt(2*np.pi)) * np.exp(-np.divide(np.power(np.log(ch)-mu1,2),2*sig1**2))

x = np.linspace(1e-3, 2400, 1000)
mu = 10
mu2 = 5
sig = 3
sig2= 1
amp = 200000
amp2 = 10000
plt.plot(x,amp*lognormal(x-50,mu,sig), color='k')
plt.plot(x,amp2*lognormal(x,mu2,sig2))
plt.axvline(np.exp(mu2-(sig2**2)), color='r')
plt.show()
```
