```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from icecream import ic
```

# Result plots

## Total energy spectrum

```python
# peaks of spectra, determined retrospectively by looking at the spectra beforehand
# i could also calculate the peaks with np.max and finding local maxima, but it doesn't change the fact that i need to look at the plots beforehand to determine the threshold for local maxima.
multiindex = [
    ["E", "E", "D1", "D1", "D2", "D2"],
    ["p1", "p2","p1", "p2", "p1", "p2"],
]
tuples = list(zip(*multiindex))
index = pd.MultiIndex.from_tuples(tuples, names=["Theoretical / Detector","Peak"])
source = pd.DataFrame({"na":[511,1275,480,1260,205,790],
                       "co":[1173.2,1332.5,1150,1300,700,810]}, index=index)
ic(source)
def peak(d,p,s):
    '''
    d: E, D1 or D2
    p: p1 or p2
    s: na or co
    example use: ic(peak('D1','p1','na'))
    '''
    return float(source.loc[(d,p),s])

# time delay
peak_0ns = 1030
peak_96ns = 1700
```

```python
def plot(data, peak1, title, peak2=None, color1='g', color2='r'):
    df = pd.read_csv(data)
    df.columns = ['counts']
    df.index.name = 'Kanal'
    df['counts'] = df.loc[df['counts'] > 0]

    plt.rc('text', usetex=True)
    plt.rc('font', family='STIXGeneral')
    sns.scatterplot(x=df.index, y=df['counts'], s=3, color='k')
    plt.title(title)
    if peak2 == None:
        plt.axvline(peak1, linewidth=.5, c=color1, label=f'peak bei {peak1}')
    else:
        plt.axvline(peak1, linewidth=.5, c=color1, label=f'peak1 bei {peak1}')
        plt.axvline(peak2, linewidth=.5, c=color2, label=f'peak2 bei {peak2}')
    
    plt.legend()
    plt.savefig(f"plots/{title}.pdf")
    plt.show()

plot('Results/energiespektrum_natrium_detektor1.asc', peak('D1','p1','na'), 'Energiegesamtspektrum von Natrium aus Detektor 1', peak2=peak('D1','p2','na'))
plot('Results/energiespektrum_natrium_detektor2.asc', peak('D2','p1','na'), 'Energiegesamtspektrum von Natrium aus Detektor 2', peak2= peak('D2','p2','na'))
plot('Results/energiespektrum_cobalt_detektor1.asc', peak('D1','p1','co'), 'Energiegesamtspektrum von Cobalt aus Detektor 1', peak2=peak('D1','p2','co'))
plot('Results/energiespektrum_cobalt_detektor2.asc', peak('D2','p1','co'), 'Energiegesamtspektrum von Cobalt aus Detektor 2', peak2=peak('D2','p2','co'))

plot('Results/energieselektion_natrium_detektor1.asc', peak('D1','p1','na'), 'Ausgewähltes Spektrum von Natrium aus Detektor 1')
plot('Results/energieselektion_natrium_detektor2.asc', peak('D2','p1','na'), 'Ausgewähltes Spektrum von Natrium aus Detektor 2')
plot('Results/energieselektion_cobalt_detektor1.asc', peak('D1','p1','co'), 'Ausgewähltes Spektrum von Cobalt aus Detektor 1', peak2=peak('D1','p2','co'))
plot('Results/energieselektion_cobalt_detektor2.asc', peak('D2','p1','co'), 'Ausgewähltes Spektrum von Cobalt aus Detektor 2', peak2=peak('D2','p2','co'))
plot('Results/energieselektion_natrium_detektor1_gamma-peak.asc', peak('D1','p2','na'), 'Ausgewähltes Spektrum von Natrium aus Detektor 1 \n beim zweiten Peak', color1='r')
'''
Despite looking at the same source and detector, the measurement of the peaks show a slightly different peak than the measurement of the complete spectra. 
I.e. the peak2 of sodium at detector 1 lies at 1260 for the complete spectrum measurement, but lies at 1250 for the selection measurement
This means that a rerun will give different results. The difference is tiny, but noticeable.
Will be noted for possible error assessment
'''

plot('Results/zeitspektrum_0nsdelay.asc', peak_0ns, 'Zeitspektrum bei 0ns Delay')
plot('Results/zeitspektrum_96nsdelay.asc', peak_96ns, 'Zeitspektrum bei 96ns Delay')
```

## Linear regression of the peaks $\&$ energy calibration

Detector|sodium 511keV|sodium 1275keV|cobalt 1173.2keV|cobalt 1332.5keV|
:-|-|-|-|-|
D1|480|1260|1150|1300|
D2|205|790|700|810|

We capitalize on the fact that the channel value of the TSCA is linear proportional to the energy values to an accuracy of 1keV ["Grundlagen_gamma_FP16" page 46]:
$$
    E_\gamma = a + b \cdot K .
$$
This means that the difference in energies should be the same as the difference in channel values to a factor $b$. We can determine the values $a$ and $b$ and calibrate our channel values accordingly.

```python
ic(source)
```

```python
delta_E_na = peak('E','p2','na') - peak('E','p1','na')

delta_D1_na = peak('D1','p2','na') - peak('D1','p1','na') 

b_D1_na = delta_D1_na / delta_E_na
a_D1_na = peak('D1','p1','na') - b_D1_na * peak('E','p1','na')

delta_D2_na = peak('D2','p2','na') - peak('D2','p1','na')
b_D2_na = delta_D2_na / delta_E_na 
a_D2_na = peak('D2','p1','na') - b_D2_na * peak('E','p1','na')


delta_E_co = peak('E','p2','co') - peak('E','p1','co')

delta_D1_co = peak('D1','p2','co') - peak('D1','p1','co') 
b_D1_co = delta_D1_co / delta_E_co
a_D1_co = peak('D1','p1','co') - b_D1_co * peak('E','p1','co')

delta_D2_co = peak('D2','p2','co') - peak('D2','p1','co')
b_D2_co = delta_D2_co / delta_E_co 
a_D2_co = peak('D2','p1','co') - b_D2_co * peak('E','p1','co')


multiindex2 = [
    ["D1", "D1", "D2", "D2"],
    ["a", "b", "a", "b"],
]
tuples = list(zip(*multiindex2))
index2 = pd.MultiIndex.from_tuples(tuples, names=["Theoretical / Detector","Peak"])
calibration = pd.DataFrame({"na":[a_D1_na,b_D1_na,a_D2_na,b_D2_na],
                       "co":[a_D1_co,b_D1_co,a_D2_co,b_D2_co]}, index=index2)

ic(calibration)

def calibrate(d,s,ch):
    '''
    d: detector, D1 or D2 
    s: source, na or co
    ch: channel value, float
    '''
    a = calibration.loc[(d,'a'),s]
    b = calibration.loc[(d,'b'),s]
    E = -(a/b) + (1/b) * ch
    return E
# ic(calibrate('D1','na',480)) # test
ch = np.linspace(0,1500,1000)
plt.plot(ch, calibrate('D1','na',ch))
plt.show()
```
