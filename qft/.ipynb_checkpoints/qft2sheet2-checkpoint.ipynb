```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
\begin{align}
F_{\text{clas}}(\beta) &= \frac{1}{\beta} \ln(\beta \omega) \\\\
F_{\text{quant}}(\beta) &= \frac{1}{\beta} \ln\left(2 \sinh\left( \frac{\beta \hbar \omega}{2} \right)\right) \\\\
F_{\text{euc}}(\beta) &= -\frac{1}{2\beta} \ln\left( \frac{m \omega}{2\pi \sinh(\omega \beta)} \right)
+ \frac{m^2 \omega x^2}{2\beta \sinh(\omega \beta)} \left( \cosh(\omega \beta) - 1 \right)
\end{align}
$$



```python
omega = 1.0  
m = 1.0  
x = 1.0 
# planck = 6.62607015e-34
planck = 1e-10
betarange = np.linspace(1e-3,0.04,int(1e5))

def F_clas(beta):
    return 1 / beta * np.log(beta * omega)

def F_quant(beta):
    return 1 / beta * np.log(2*np.sinh(beta * planck * omega / 2))

def F_euc(beta):
    scl = -1 / (2 * beta) * np.log(m * omega / (2 * np.pi * np.sinh(omega * beta)))
    sy = (m**2 * omega * x**2) / (2 * beta * np.sinh(omega * beta)) * (np.cosh(omega * beta) - 1)
    return scl + sy

plt.figure(figsize=(9, 5))
plt.plot(betarange, F_clas(betarange), label=r'$F_{clas}(\beta)$', color='b' )
plt.plot(betarange, F_quant(betarange), label=r'$F_{quant}(\beta)$', color='g')
plt.plot(betarange, F_euc(betarange), label=r'$F_{euc}(\beta)$', color='r')
plt.xlabel(r'$\beta$', fontsize=14)
plt.ylabel(r'$F$', fontsize=14)
plt.title(r'F - $\beta$', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
betarange = np.linspace(0.5,100,int(1e5))

plt.figure(figsize=(9, 5))
plt.plot(betarange, F_clas(betarange), label=r'$F_{clas}(\beta)$', color='b' )
plt.plot(betarange, F_quant(betarange), label=r'$F_{quant}(\beta)$', color='g')
plt.plot(betarange, F_euc(betarange), label=r'$F_{euc}(\beta)$', color='r')
plt.xlabel(r'$\beta$', fontsize=14)
plt.ylabel(r'$F$', fontsize=14)
plt.title(r'F - $\beta$', fontsize=16)
plt.ylim(-15, 2)
plt.legend()
plt.grid(True)
plt.show()
```
